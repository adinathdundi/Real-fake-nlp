```python
#reading the data
data_set_fake = pd.read_csv('/kaggle/input/fake-and-real-news-dataset/Fake.csv',parse_dates = ['date'])
data_set_true = pd.read_csv('/kaggle/input/fake-and-real-news-dataset/True.csv',parse_dates = ['date'])
```

```python
#labelling the data
data_set_fake['label'] = 'fake'
data_set_true['label'] = 'true'
```

```python
#merging the real & fake data
data_set_full = pd.concat([data_set_fake,data_set_true],axis = 0)
#shuffling the data 
data_set_full = data_set_full.sample(frac=1).reset_index(drop=True)

data_set_full.head()
```

```python
#tokenizing my docs
import spacy
from spacy.tokenizer import Tokenizer

#loading the model
nlp = spacy.load("en_core_web_sm")
#creating a bag of words vector
textcat = nlp.create_pipe("textcat",config={"exclusive_classes":True,"architecture":"bow"})
#adding the model to the pipeline
nlp.add_pipe(textcat)
```

```python
#adding labels to the data
textcat.add_label("fake")
textcat.add_label("true")
```

```python
train_texts = data_set_full['title']
train_labels = [{'cats':{'fake':label == 'fake','true':label == 'true'}} for label in data_set_full['label']]
train_data = list(zip(train_texts, train_labels))
train_data[:3]
```
